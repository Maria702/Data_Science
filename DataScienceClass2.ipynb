# data.mean()
   - Return the mean of the values for the requested axis.

```python
data['Attack'].mean()
```

```python
#  'Attack_high_low' 
# # if attack is more than 79 its called Attack_high
if attack is less than 79 its called Attack_low
```

 # Add a Column..

```python
attack_mean = data['Attack'].mean()

def set_attack(val):
    if val < attack_mean:
        return 'Attack High'
    else:
        return 'Attack Low'
```

```python
# Apply a function along an axis of the DataFrame.

data['Attack_High_Low']= data['Attack'].apply(set_attack)
data
```
